```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

# Model
model = LinearRegression()

# Train
model.fit(X, y)

# Parameters
m = model.coef_[0]
b = model.intercept_

print("Slope (m):", m)
print("Intercept (b):", b)

# Predictions
y_pred = model.predict(X)
print("Predictions:", y_pred)

# New prediction
new_x = np.array([[6]])
print("Prediction for x=6:", model.predict(new_x)[0])
```
