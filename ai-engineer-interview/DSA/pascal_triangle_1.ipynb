Program to generate Pascal's Triangle


Problem Statement: Write a program to generate Pascal's triangle. In Pascal’s triangle, each number is the sum of the two numbers directly above it as shown in the figure below:





Input: N = 5, r = 5, c = 3 

Output: Element at position (r, c): 6

N-th row of Pascal’s triangle: 1 4 6 4 1

First n rows of Pascal’s triangle:

1 

1 1 

1 2 1 

1 3 3 1 

1 4 6 4 1  

Explanation: Pascal triangle for first 5 rows is shown above.

Input: N = 1, r = 1, c = 1

Output: Element at position (r, c): 1

N-th row of Pascal’s triangle: 1

First n rows of Pascal’s triangle:

1  

Explanation: N = 1 is the base case fof a pascal's triangle.

```python
###### What is Pascal’s Triangle?

# Pascal’s Triangle is a triangular pattern where:

# First row is always: 1
# Every number inside = sum of the two numbers just above it

# Example:
#         1
#       1   1
#     1   2   1
#   1   3   3   1
# 1   4   6   4   1


###### Key Rule (Most Important)

# Each element is formed as:

# value = above_left + above_right

# Example:

# 6 = 3 + 3
# 4 = 1 + 3

# Edges are always 1 because they don’t have two numbers above.

##### Important Insight (Math View)

# Pascal’s Triangle is actually based on combinations:

# We use:

# C(n,r) = n! / r!(n-r)!

 
##### Indexing Trick (VERY IMPORTANT)

# In coding:

# Rows start from 0
# Columns start from 0

# So:

# Row = r - 1
# Column = c - 1



##### Example Explained
# Input:
# N = 5, r = 5, c = 3

# Convert to 0-based:

# Row = 4
# Column = 2

# Now compute:

# C(4, 2) = 4! / (2! * 2!)
#         = 24 / (2 * 2)
#         = 6

# Output = 6


# N-th Row Example

# For N = 5:

# We compute:

# C(4,0) C(4,1) C(4,2) C(4,3) C(4,4)

# Which becomes:

# 1 4 6 4 1
```

```python
# Solution 

triangle = []
for i in range(5):        # runs n times
    row = [1] * (i + 1)   # # work depends on i
    for j in range(1, i): ## runs i times
        row[j] = triangle[i - 1][j - 1] + triangle[i - 1][j]
    triangle.append(row)

print(triangle)

```

```python
# Intuition (Best way to remember)

# Think like this:

# “Each position has two parents above it”

# left parent → j-1
# right parent → j
# So child = sum of parents
```
