Kadane's Algorithm : Maximum Subarray Sum in an Array

Problem Statement: Given an integer array nums, find the subarray with the largest sum and return the sum of the elements present in that subarray.

A subarray is a contiguous non-empty sequence of elements within an array.

Example 1:

Input:

nums = [2, 3, 5, -2, 7, -4]  

Output:

15  

Explanation:

The subarray from index 0 to index 4 has the largest sum = 15, which is the maximum sum of any contiguous subarray.

Example 2:

Input:

nums = [-2, -3, -7, -2, -10, -4]  

Output:

-2  

Explanation:

The largest sum is -2, which comes from taking the element at index 0 or index 3 as the subarray. Since all numbers are 
negative, the subarray with the least negative number gives the largest sum.

```python
# Core Idea (Simple Intuition)

# At every index, you decide:

# “Should I continue the current subarray OR start fresh from here?”



# Rule
# If current sum becomes negative, discard it (start fresh)
# Keep track of maximum sum seen so far


# Algorithm (Step-by-step)
# Initialize:
# current_sum = 0
# max_sum = nums[0]
# Loop through array:
# Add element to current_sum
# Update max_sum = max(max_sum, current_sum)
# If current_sum < 0, reset it to 0
```

```python
arr = [2,3,5,-2,7,-4 ]
current_sum = 0 
max_sum = arr[0]

for num in arr: 
    current_sum += num

    max_sum = max(max_sum,current_sum)

    if current_sum < 0:
        current_sum = 0

print(max_sum)
```
