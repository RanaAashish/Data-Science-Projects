Dutch National Flag Algorithm

Sort an array of 0s, 1s and 2s


Problem Statement: Given an array nums consisting of only 0, 1, or 2. Sort the array in non-decreasing order. The sorting must be done in-place, without making a copy of the original array.

Input: nums = [1, 0, 2, 1, 0]

Output: [0, 0, 1, 1, 2]

Explanation: The nums array in sorted order has 2 zeroes, 2 ones and 1 two

Input: nums = [0, 0, 1, 1, 1]

Output: [0, 0, 1, 1, 1]

Explanation: The nums array in sorted order has 2 zeroes, 3 ones and zero twos.

```python
# Idea (Very Important)

# You only have 3 numbers: 0, 1, 2

# So instead of sorting like normal (O(n log n)), we can do it in one pass (O(n)).

# We divide the array into 3 regions:

# [ 0s | 1s | unknown | 2s ]

# We use 3 pointers:

# low → where next 0 should go
# mid → current element
# high → where next 2 should go



# Rules (Core Logic)

# While mid <= high:

# 1. If element is 0

# Swap with low
# Move both low++ and mid++

# 2. If element is 1

# Correct place already
# Just do mid++

# 3. If element is 2

# Swap with high
# Do high--
# DO NOT move mid (important!)


# Why not move mid when it's 2?

# Because the swapped element from high could be 0, 1, or 2
# So we need to check it again.
```

```python
nums = [1,0,2,1,0]
low = 0
mid = 0
high = len(nums) - 1

while mid <= high:
    if nums[mid] == 0:
        nums[low], nums[mid] = nums[mid], nums[low]
        low += 1
        mid += 1
    
    elif nums[mid] == 1:
        mid += 1

    else:  # nums[mid] == 2
        nums[mid], nums[high] = nums[high], nums[mid]
        high -= 1
print(nums)
```
