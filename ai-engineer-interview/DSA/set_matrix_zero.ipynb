1. Set Matrix Zero

Problem Statement: Given a matrix if an element in the matrix is 0 then you will have to set its entire column and row to 0 and 

then return the matrix..

Examples: 

Input: matrix=[[1,1,1],[1,0,1],[1,1,1]]

Output: [[1,0,1],[0,0,0],[1,0,1]]

Explanation: Since matrix[2][2]=0.Therfore the 2nd column and 2nd row wil be set to 0.

Input: matrix=[[0,1,2,0],[3,4,5,2],[1,3,1,5]]

Output:[[0,0,0,0],[0,4,5,0],[0,3,1,0]]

Explanation:Since matrix[0][0]=0 and matrix[0][3]=0. Therefore 1st row, 1st column and 4th column will be set to 0


```python
# Solution i came up with  

input_matrix = [
    [1,1,1],
    [1,0,1],
    [1,1,1],
    ]

# Solution - 
# - let's make a another matrix 
# - Go to each element one by one 
# - Find 0
# - If there is 0 , in rown_number[element_number], then put that row number to 0 
# - consider that element number as a column number and put that to 0 


output_matrix = [
    [1,0,1],
    [0,0,0],
    [1,0,1]
    ]
```

```python
n = len(input_matrix)
print(n)
print(input_matrix[0])
m = len(input_matrix[0])
print(m)
```

```python
rows = [False] * n
columns = [False] * m 
print(columns)
```

```python
rows = [False] * n
columns = [False] * m 
for i in range (n):
    for j in range (m): 
        if input_matrix[i][j] == 0 : 
            rows[i] = True
            columns[j]= True 

print(columns)
print(rows)
```

```python
for i in range(n): 
    for j in range(m): 
        if rows[i] or columns[j]: 
            input_matrix[i][j] = 0 
```
